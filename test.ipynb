```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt

# Simulating a small dataset for demonstration purposes
np.random.seed(42)
data = pd.DataFrame({
    'tenure': np.random.randint(1, 72, size=100),          # Numerical feature
    'MonthlyCharges': np.random.uniform(20, 120, 100),    # Numerical feature
    'Contract': np.random.choice(['Month-to-Month', 'One Year', 'Two Year'], size=100),  # Categorical feature
    'Churn': np.random.choice(['Yes', 'No'], size=100)    # Target variable
})

# Create a "Distribution of Features" plot
fig, axes = plt.subplots(2, 2, figsize=(12, 10))
axes = axes.flatten()

# Distribution of numerical feature: tenure
sns.histplot(data=data, x='tenure', kde=True, bins=20, ax=axes[0], color='blue')
axes[0].set_title('Distribution of Tenure')

# Distribution of numerical feature: MonthlyCharges
sns.histplot(data=data, x='MonthlyCharges', kde=True, bins=20, ax=axes[1], color='green')
axes[1].set_title('Distribution of Monthly Charges')

# Distribution of categorical feature: Contract
sns.countplot(data=data, x='Contract', ax=axes[2], palette='viridis')
axes[2].set_title('Distribution of Contract Types')

# Distribution of categorical feature: Churn
sns.countplot(data=data, x='Churn', ax=axes[3], palette='coolwarm')
axes[3].set_title('Distribution of Churn')

# Adjust layout
plt.tight_layout()
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt

# Example dataset
data = pd.DataFrame({
    'tenure': [1, 12, 24, 36, 48, 60, 72],
    'MonthlyCharges': [20.5, 35.7, 50.3, 75.6, 100.2, 120.9, 65.4],
    'TotalCharges': [102, 358, 803, 2200, 4800, 7200, 5000]
})

# Calculate basic statistics
statistics = data.agg(['mean', 'median', lambda x: x.mode()[0]])
statistics.index = ['Mean', 'Median', 'Mode']

# Plotting the table
fig, ax = plt.subplots(figsize=(6, 3))
ax.axis('off')  # Remove axes
table = ax.table(cellText=statistics.values,
                 colLabels=statistics.columns,
                 rowLabels=statistics.index,
                 cellLoc='center',
                 loc='center')
table.auto_set_font_size(False)
table.set_fontsize(10)
table.scale(1.2, 1.5)  # Adjust size
plt.title("Basic Statistics (Mean, Median, Mode)")
plt.show()
```

```python
import numpy as np

# Transpose statistics for plotting
statistics_transposed = statistics.T
statistics_transposed.plot(kind='bar', figsize=(8, 5))

# Customizing the plot
plt.title('Basic Statistics (Mean, Median, Mode)', fontsize=16)
plt.ylabel('Value', fontsize=12)
plt.xlabel('Features', fontsize=12)
plt.xticks(rotation=0)
plt.legend(title='Statistic')
plt.tight_layout()
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt

def visualize_basic_statistics(data):
    """
    Visualizes basic statistics (mean, median, mode) for numeric and categorical features
    using a bar chart.
    
    Parameters:
        data (pd.DataFrame): The dataset to analyze.
    """
    # Dictionary to store statistics for each column
    statisticsNumeric = {}
    statisticsCategorical = {}
    
    for column in data.columns:
        if pd.api.types.is_numeric_dtype(data[column]):
            # If numeric, calculate mean, median, and mode
            statisticsNumeric[column] = {
                'Mean': data[column].mean(),
                'Median': data[column].median(),
                'Mode': data[column].mode()[0]
            }
        elif data[column].dtype == 'object':
            # If categorical, calculate only mode
            statisticsCategorical[column] = {
                'Mode': data[column].mode()[0]
            }

    # Convert dictionary to a DataFrame for plotting
    stats_df = pd.DataFrame(statistics).T

    # Plot bar chart
    stats_df.plot(kind='bar', figsize=(10, 6), width=0.8)
    plt.title('Basic Statistics (Mean, Median, Mode)', fontsize=16)
    plt.ylabel('Value', fontsize=12)
    plt.xlabel('Features', fontsize=12)
    plt.xticks(rotation=45, ha='right')
    plt.legend(title='Statistic')
    plt.tight_layout()
    plt.show()
```

```python
# Example dataset
data = pd.DataFrame({
    'tenure': [1, 12, 24, 36, 48, 60, 72],            # Numeric
    'MonthlyCharges': [20.5, 35.7, 50.3, 75.6, 100.2, 120.9, 65.4],  # Numeric
    'Contract': ['Month-to-Month', 'One Year', 'One Year', 'Two Year', 
                 'Month-to-Month', 'Month-to-Month', 'Two Year'],     # Categorical
    'Churn': ['Yes', 'No', 'No', 'Yes', 'Yes', 'No', 'No']           # Categorical
})

# Visualize statistics
visualize_basic_statistics(data)
```
